```python
a=int(input())
def so_nguyen_to(a):
    if a<=1:
        return 'không là số nguyên tố'
    for i in range(2,int(a**0.5)+1):
        if a%i==0:
            return 'không số nguyên tố'
    
    return "là số nguyên tố"
print(so_nguyen_to(a))
    
```

W5A4
VIết hàm kiểm tra một số có phải số hoàn hảo hay không, xử lý trên dữ liệu kiểu số nguyên. 

```python
n=int(input())
def kiem_tra_so_hoan_hao(n):
    if n <= 1:
        return False
    tong_uoc = 0
    for i in range(1, n // 2 + 1):
        if n % i == 0:
            tong_uoc += i
    return tong_uoc == n
print(kiem_tra_so_hoan_hao(n))
    
```

Viết hàm đưa vào một list số nguyên và một số nguyên dương k. Hãy tìm và trả về vị trí của phần tử 
đầu tiên có giá trị k trong list số nguyên, nếu không có thì trả về -1 

```python
def tim_vi_tri_dau_tien(danh_sach, k):
    for vi_tri in range(len(danh_sach)):
        if danh_sach[vi_tri] == k:
            return vi_tri
    return -1

danh_sach_so = [10, 5, 7, 3, 5, 8]
so_can_tim = 5

ket_qua = tim_vi_tri_dau_tien(danh_sach_so, so_can_tim)
print("Vị trí tìm thấy là:", ket_qua)
```

liệu bị lồng thêm vào, hoặc bị xóa đi, phương pháp đo lường phức tạp hơn, như khoảng cách 
Levenshtein (Levenshtein distance) là một phương pháp có tác dụng và thích hợp hơn.

W5A15 Viết hàm is_right_triangle(a, b, c) trả về True nếu ba số có thể là ba cạnh của tam giác vuông. 

```python
def is_right_triangle(a, b, c):
    if a <= 0 or b <= 0 or c <= 0:
        return False
        
    if a + b <= c or a + c <= b or b + c <= a:
        return False
        
    if a**2 + b**2 == c**2 or a**2 + c**2 == b**2 or b**2 + c**2 == a**2:
        return True
        
    return False

print(is_right_triangle(3, 4, 5))
print(is_right_triangle(5, 12, 13))
print(is_right_triangle(1, 2, 3))
```
