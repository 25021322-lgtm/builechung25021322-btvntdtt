W7A1 
Viết hàm tìm kiếm nhị phân 
trên một danh sách đã 
được sắp xếp tăng dần. 
Trả về chỉ số của phần tử 
hoặc -1 nếu không tìm 
thấy. mẫu hàm: def 
binary_search(arr, left, 
right, target): 

```python
def binary_search(ds, num):
    left = 0
    right = len(ds) - 1
    while left <= right:
        mid = (left + right)//2
        if ds[mid] == num:
            return mid
        elif ds[mid] < num:
            left = mid + 1
        else:
            right = mid - 1
    else:
        return -1

ds = list(map(int, input().split()))
num = int(input())
print(binary_search(ds, num))
```

W7A3 
Viết hàm thực hiện thuật 
toán bubble sort, Selection 
Sort để sắp xếp danh sách 
theo thứ tự tăng dần. Sắp 
xếp nổi bọt (Bubble Sort);Sắp xếp chọn (Selection 
Sort) 

```python
def Bubble_Sort(arr):
    for i in range(len(arr)):
        for i in range(0, len(arr) - i - 1):
            if arr[i] > arr[i + 1]:
                arr[i], arr[i + 1] = arr[i + 1], arr[i]
    return arr

def Selection_Sort(arr):
    for i in range(len(arr)):
        minidx = i
        for j in range(i + 1, len(arr)):
            if arr[j] < arr[minidx]:
                minidx = j
        arr[i], arr[minidx] = arr[minidx], arr[i]
    return arr

print(Bubble_Sort(arr = list(map(int, input().split()))))
print(Selection_Sort(arr = list(map(int, input().split()))))
```

W7A4 
Viết hàm tìm phần tử có số 
lần xuất hiện nhiều nhất 
trong danh sách. Nếu có 
nhiều phần tử cùng số lần, 
trả về phần tử xuất hiện 
trước.

```python
def Performance(lst):
    number = 0
    result = 0
    for i in set(lst):
        dem = 0
        for j in lst:
            if i == j:
                dem += 1
        if dem > result:
            result = dem
            number = i
    return f"{number} suất hiện nhiều nhất, sớm nhất {result} lần"

lst = list(map(int, input().split()))
print(Performance(lst))
```

Cho danh sách các số 
nguyên và một số X, đếm 
tất cả các cặp (a, b) sao 
cho a + b = X. c 

```python
def findsum(lst, x):
    dem = 0
    for i in range(len(lst)):
        for j in range(i + 1, len(lst)):
            if lst[i] + lst[j] == x:
                dem += 1
    return dem

lst = list(map(int, input().split()))
x = int(input())
print(findsum(lst, x))
```

Bạn được cung cấp: một 
danh sách các khoảng thời 
gian interval = [[l1, r1], [l2, 
r2], ...] và một danh sách 
các truy vấn queries = [q1, 
q2, ...]  
Với mỗi truy vấn q, hãy tìm 
độ dài nhỏ nhất của 
khoảng thời gian [li, ri] sao 
cho li ≤ q ≤ ri.  
Nếu không tồn tại khoảng 
thời gian nào như vậy, hãy 
trả về -1 cho truy vấn đó. 
solve in < O(m.n), hint: 
binary search in the 
queries list 

```python
interval = [[1, 3], [5, 7], [9, 12]]
queries =  [2, 4, 5, 8, 10] 
dem = 0
result = [float('inf')] * len(queries)
for q in queries:
    for i in interval:
        if i[0] <= q <= i[1]:
            dem = i[1] - i[0] + 1
            if dem < result[queries.index(q)]:
                result[queries.index(q)] = dem
print([-1 if i == float('inf') else i for i in result])
```

Dãy hài hòa là một dãy 
thỏa mãn điều kiện giá trị 
lớn nhất và giá trị nhỏ nhất 
của dãy khác nhau đúng 
bằng \( 1 \). 
Viết hàm trả về độ dài của 
dãy hài hòa con dài nhất có 
thể có trong mảng giá trị 
nguyên \( arr \) có \( n \) 
phần tử 
Ví dụ: 
Input: 1 3 2 2 5 2 3 7 
Output: 5 - Dãy hài hòa con 
dài nhất là [3,2,2,2,3] 

```python
def DayHaiHoa(n, arr):
    dem = 0
    lenmax = 0
    for i in set(arr):
        dem = arr.count(i) + arr.count(i + 1)
        if dem > lenmax:
            lenmax = dem
    return lenmax

n = int(input())
arr = list(map(int, input().split()))
print(DayHaiHoa(n, arr))
```
