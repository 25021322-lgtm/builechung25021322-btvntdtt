Nhập vào một dòng gồm các số nguyên. Hãy loại bỏ phần tử trùng nhưng 
giữ nguyên thứ tự xuất hiện đầu tiên, rồi in ra list kết quả. 
Yêu cầu in ra đúng dạng list Python, ví dụ [1, 2, 3] 

```python

numbers = list(map(int, input().split()))
result = []
seen = set()
for num in numbers:
    if num not in seen:
        result.append(num)
        seen.add(num)
print(result)
```

Nhập một dòng gồm các cặp key:value cách nhau bởi dấu cách. Tạo 
dictionary ánh xạ mỗi key -> list các value theo thứ tự xuất hiện. Nếu key 
lặp lại, thêm vào list của key đó 

```python
ds = list(input().split())

key = [ds[i][:ds[i].index(':')] for i in range(len(ds))]
value = [ds[i][ds[i].index(':') + 1:] for i in range(len(ds))]
dictionalry = {}

for i in range(len(value)):
    if (key[i] in dictionalry) == False:
        dictionalry[key[i]] = []
    dictionalry[key[i]].append(value[i])
print(dictionalry)
```

Viết chương trình nhập vào hai dictionary. Sau đó gộp lại thành một 
dictionary mới trong đó nếu một key xuất hiện trong cả hai dict, giá trị sẽ 
được cộng lại, nếu chỉ có trong một dict thì giữ nguyên. In ra dictionary mới 
theo thứ tự từ điển tăng dần của key.  

```python
print("--- NHẬP TỪ ĐIỂN 1 ---")
so_luong_1 = int(input("Nhập số lượng phần tử của từ điển 1: "))
tu_dien_1 = {}
for i in range(so_luong_1):
    khoa = input("Nhập khóa: ")
    gia_tri = int(input("Nhập giá trị: "))
    tu_dien_1[khoa] = gia_tri

print("--- NHẬP TỪ ĐIỂN 2 ---")
so_luong_2 = int(input("Nhập số lượng phần tử của từ điển 2: "))
tu_dien_2 = {}
for i in range(so_luong_2):
    khoa = input("Nhập khóa: ")
    gia_tri = int(input("Nhập giá trị: "))
    tu_dien_2[khoa] = gia_tri

tu_dien_gop = {}

for khoa, gia_tri in tu_dien_1.items():
    tu_dien_gop[khoa] = gia_tri

for khoa, gia_tri in tu_dien_2.items():
    if khoa in tu_dien_gop:
        tu_dien_gop[khoa] += gia_tri
    else:
        tu_dien_gop[khoa] = gia_tri

tu_dien_sap_xep = dict(sorted(tu_dien_gop.items()))

print("--- KẾT QUẢ TỪ ĐIỂN MỚI ---")
print(tu_dien_sap_xep)
```

Viết chương trình nhập vào một danh sách các số nguyên, và số nguyên 
dương k. In ra list gồm các cặp số khác nhau có tổng bằng k. Mỗi cặp là 
một tuple, hai số trong tuple có thể giống nhau, thứ tự xuất hiện trong danh 
sách của số thứ nhất nhỏ hơn thứ tự của số thứ hai. Các tuple trong list được 
sắp xếp theo thứ tự tăng dần của phần tử đầu tiên trong tuple.

```python
ds = list(map(int, input().split()))
k = int(input())
kq = []
for i in range(len(ds)):
    for j in range(i + 1, len(ds)):
        if ds[i] + ds[j] == k and (ds[i], ds[j]) not in kq:
            kq.append((ds[i], ds[j]))
kq.sort()
print(kq)
```

```python
ds = list(map(int, input().split()))
n = float('inf')
d = 0
for i in set(ds):
    if ds.count(i) > d:
        n = i
        d = ds.count(i)
    elif ds.count == d and i < n:
        n = i
        d = ds.count(i)
print(n)
```

W6A13 [Đảo ngược key và value của dictionary] Nhập vào một dictionary có value là duy nhất (không trùng nhau). Hãy tạo một dictionary mới bằng cách đảo ngược key và value của dictionary ban đầu. In ra dictionary mới.

```python
ds = list(input().split())
key = [ds[i] for i in range(len(ds)) if i % 2 != 0]
value = [ds[j] for j in range(len(ds)) if j % 2 == 0]
dictionalry = {}
for k in range(len(key)):
    dictionalry[key[k]] = value[k]
print(dictionalry)
```

W6A15 [Lọc dictionary theo giá trị] Nhập vào một dictionary có key là chuỗi và value là số nguyên, và một số nguyên k ở dòng tiếp theo. Hãy tạo một dictionary mới chỉ chứa các cặp key value từ dictionary ban đầu mà có value lớn hơn k. In ra dictionary mới.

```python
ds = list(input().split())
k = int(input())
key = [ds[i] for i in range(len(ds)) if i % 2 == 0]
value = [int(ds[j]) for j in range(len(ds)) if j % 2 != 0]
dictionalry = {}

for i in range(len(value)):
    if value[i] > k:
        dictionalry.update({key[i] : value[i]})
print(dictionalry)
```
