Nhập vào mảng vuông 2 chiều n hàng, n cột, in ra đường chéo chính và 
đường chéo phụ của mảng

```python
n = int(input())
matrix = []
for i in range(n):
    matrix.append([])
    for j in range(n):
        matrix[i].append(input())
print("Các phần tử trên đường chéo chính:", end = "")
for i in range(n):
    print(matrix[i][i], end = " ")
print("Các phần tử trên đường chéo phụ:", end = "")
for i in range(n - 1, -1, -1):
    print(matrix[n-i-1][i], end = " ")
```

Viết hàm đưa vào một list số nguyên và một số nguyên dương k. Hãy tìm và 
trả về vị trí của phần tử đầu tiên có giá trị k trong list, nếu không có thì trả về -1

W6A21 
Nhập một dòng gồm các chuỗi (tách bởi dấu cách). Tạo dictionary ánh xạ 
chuỗi -> tuple các vị trí (0-based) mà chuỗi đó xuất hiện 

```python
dong_nhap_vao = input("Nhập các chuỗi cách nhau bởi dấu cách: ")
danh_sach_chuoi = dong_nhap_vao.split()

tu_dien_vi_tri = {}

for vi_tri in range(len(danh_sach_chuoi)):
    chuoi = danh_sach_chuoi[vi_tri]
    if chuoi in tu_dien_vi_tri:
        tu_dien_vi_tri[chuoi] = tu_dien_vi_tri[chuoi] + (vi_tri,)
    else:
        tu_dien_vi_tri[chuoi] = (vi_tri,)

print("Kết quả ánh xạ là:")
print(tu_dien_vi_tri)
```

W6A22 
Viết hàm đưa vào 1 list số nguyên, tìm và trả về vị trí có giá trị lớn nhất 
trong list 

```python
def tim_vi_tri_lon_nhat(danh_sach):
    if len(danh_sach) == 0:
        return -1
        
    vi_tri_max = 0
    
    for i in range(1, len(danh_sach)):
        if danh_sach[i] > danh_sach[vi_tri_max]:
            vi_tri_max = i
            
    return vi_tri_max

day_so = [10, 25, 7, 42, 36, 42]
ket_qua = tim_vi_tri_lon_nhat(day_so)
print("Vị trí của phần tử lớn nhất là:", ket_qua)
```

W6A24 Viêết hàm đẩy tất cả các số 0 về đầu (về cuối), các số khác vẫn đúng vị trí trước sau: 0 2 3 2 4 0 0 0 9 => 0 0 0 2 3 2 4 9

```python
def lendau(ds):
    ds1 = []
    ds2 = []
    for i in ds:
        if i == 0:
            ds1.append(0)
        else:
            ds2.append(i)
    return ds1 + ds2
print(lendau([0, 2, 3, 2, 4, 0, 0, 0, 9]))
```

W6A26 Tính tổng các phần tử cực đại địa phương trong một danh sách, không tính 2 phần tử đầu và cuối

```python
def tongcucdai(ds):
    s = 0
    for i in range(1, len(ds)-1):
        if ds[i] > ds[i+1] and ds[i] > ds[i-1]:
            s += ds[i]
    return s
print(tongcucdai([1, 2, 1, 4, 5, 4, 7, 2]))
```

W6A28 Cho một dãy các số nguyên, kiểm tra xem nó là dãy tăng thực sự hay giảm thực sự, hay vô danh (không tăng, không giảm)

```python
def tanggiam(ds):
    if ds[0] < ds[1]:
        for i in range(1, len(ds) -1 ):
            if ds[i] >= ds[i+1]:
                return "Dãy vô danh"
        else:
            return "Dãy tăng"
    elif ds[0] > ds[1]:
        for i in range(1, len(ds) - 1):
            if ds[i] <= ds[i+1]:
                return "Dãy vô danh"
        else:
            return "Dãy giảm"
    else:
        return "Dãy vô danh"
tanggiam([5, 4, 3, 2, 1])
```
