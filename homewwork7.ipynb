Cho một mảng số nguyên không âm. Gọi mảng đó là "mảng đẹp" nếu 
mỗi phần tử x trong mảng xuất hiện đúng x lần. 
Kiểm tra xem mảng có phải là mảng đẹp hay không. Nếu không hãy xóa 
đi y số (y là nhỏ nhất có thể) để mảng còn lại thành mảng đẹp. 

```python
lst = list(map(int, input().split()))
test = True
for i in set(lst):
    if i != lst.count(i):
        test = False
        break
else:
    print("Mảng đẹp")

if test == False:
    for i in set(lst) and i == 0:
        if i > lst.count(i):
            lst = [x for x in lst if x != i]
        elif i < lst.count(i):
            while i != lst.count(i):
                lst.remove(i)
    print(lst)
```

Tìm phần tử xuất hiện nhiều hơn n/2 lần trong mảng (nếu có) với độ 
phức tạp thời gian là O(n), độ phức tạp bộ nhớ là O(1), - Boyer–Moore. 
Nếu không có phần tử nào thỏa mãn thì không in gì

```python
lst = list(map(str, input().split()))
l = list(set(lst))
dem = [0] * len(l)
for i in lst:
    dem[l.index(i)] += 1
for i in range(len(dem)):
    if dem[i] > len(lst)/2:
        print(l[i])
```

Sắp xếp có điều kiện: Số chẵn đứng trước, số lẻ đứng sau, chẵn tăng 
dần, lẻ giảm dần; ;  

```python
def sap_xep(lst):
    le=[]
    chan=[]
    for i in lst:
        if i%2==0:
            chan.append(i)
        else:
            le.append(i)
    chan.sort()
    le.sort(reverse=True)
    ket_qua=chan + le
    return ket_qua
lst=[1, 2, 3, 4, 4, 1, 9, 5, 8]
print(sap_xep(lst))
```

Cho N tên sinh viên gồm cả họ và tên, tìm xem tên nào hay gặp nhất. 
Nếu có nhiều tên cùng số lần gặp thì lấy tên có thứ tự từ điển nhỏ nhất 

```python
def tim_ten_pho_bien_nhat(danh_sach_ten):
    if not danh_sach_ten:
        return None
    
    dem_ten = {}
    for ten in danh_sach_ten:
        dem_ten[ten] = dem_ten.get(ten, 0) + 1
    
    so_lan_lon_nhat = max(dem_ten.values())
    
    ung_vien = []
    for ten, so_lan in dem_ten.items():
        if so_lan == so_lan_lon_nhat:
            ung_vien.append(ten)
    
    ung_vien.sort()
    
    return ung_vien[0]

ds_sinh_vien = ["Bui Le Chung", "Nguyen Van A", "Bui Le Chung", "An Nhien", "An Nhien"]
print(tim_ten_pho_bien_nhat(ds_sinh_vien))
```

W7A24 Cho xâu kí tự, xóa các từ giống nhau liền kề, chỉ giữ lại từ đầu tiên 

```python
def xoa_tu_trung_lap(s):
    tu_list = s.split()
    
    if not tu_list:
        return ""
    
    ket_qua = [tu_list[0]]
    
    for i in range(1, len(tu_list)):
        if tu_list[i] != tu_list[i-1]:
            ket_qua.append(tu_list[i])
            
    return " ".join(ket_qua)

xau_vao = "hoc hoc nua nua nua hoc mai"
xau_ra = xoa_tu_trung_lap(xau_vao)

print("- Xâu ban đầu:", xau_vao)
print("- Xâu sau khi xử lý:", xau_ra)
```

```python
def xu_ly_so_trong_van_ban(van_ban):
    danh_sach_so = []
    so_tam_thoi = ""
    
    for ky_tu in van_ban:
        if ky_tu.isdigit():
            so_tam_thoi += ky_tu
        else:
            if so_tam_thoi != "":
                danh_sach_so.append(int(so_tam_thoi))
                so_tam_thoi = ""
    
    if so_tam_thoi != "":
        danh_sach_so.append(int(so_tam_thoi))
    
    if not danh_sach_so:
        return 0, 0, []

    tong = sum(danh_sach_so)
    trung_binh = tong / len(danh_sach_so)
    
    return tong, trung_binh, danh_sach_so

doan_van = "Trong 2 ngày, nhóm đã hoàn thành 10 bài tập và đạt 18 điểm."
tong_so, trung_binh_so, danh_sach = xu_ly_so_trong_van_ban(doan_van)

print("- Văn bản:", doan_van)
print("- Danh sách các số:", danh_sach)
print("- Tổng các số:", tong_so)
print("- Trung bình cộng:", trung_binh_so)
```

Bà ̀ i toán Tìm dãy các hợp số dài nhất liên tiếp không chứa số nguyên 
nào <= N (N là số nguyên dương cho trước). Nếu có nhiều dãy cùng độ

```python
numbers = input().strip()
n = int(input())
def songuyento(number):
    for i in range(2, number):
        if number % i == 0:
            return True
    else:
        return False
hopso = ""
s = ""
d = 0
for i in numbers:
    if int(i) > n and i != 1 and songuyento(int(i)) == True:
        s += i
    else:
        if len(s) > d:
            d = len(s)
            hopso = s
            s = ""
print(hopso)
```

Tìm Dãy con tăng dần liên tục dài nhất của một danh sách nhập vào từ 
bàn phím. Lấy dãy đầu tiên nếu có nhiều dãy có số phần tử bằng nhau 

```python
def tim_day_con_dai_nhat():
    nhap_lieu = input("Nhập các số cách nhau bằng khoảng trắng: ")
    a = [int(x) for x in nhap_lieu.split()]
    
    if not a:
        return []

    max_day = []
    day_hien_tai = [a[0]]

    for i in range(1, len(a)):
        if a[i] > a[i-1]:
            day_hien_tai.append(a[i])
        else:
            if len(day_hien_tai) > len(max_day):
                max_day = day_hien_tai
            day_hien_tai = [a[i]]
    
    if len(day_hien_tai) > len(max_day):
        max_day = day_hien_tai
        
    return max_day

ket_qua = tim_day_con_dai_nhat()
print("- Dãy con tăng dần liên tục dài nhất đầu tiên là:", ket_qua)
```

W7A30 
Tìm Dãy con tăng dần KHÔNG liên tục dài nhất của một danh sách nhập 
vào từ bàn phím  

```python
lst = list(map(int, input().split()))
result = []

for i in range(len(lst)):
    x = [lst[i]]
    for j in range(i + 1, len(lst)):
        if lst[j] > x[-1]:
            x.append(lst[j])
        elif lst[j] < x[-1] and len(x) >= 2:
            if lst[j] > x[-2]:
                x[-1] = lst[j]
    if len(x) > len(result):
        result = x

print(result)
```
