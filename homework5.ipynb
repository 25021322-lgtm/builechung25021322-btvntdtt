W5A16 Viết hàm kiểm tra một số có phải số Amstronhg hay không? Hàm trả về kiểu bool, có tham số là một số nguyên x

```python
def Amstronhg(n):
    k = str(n)
    number = 0
    for i in k:
        number += int(i)**len(k)
    if number == n:
        return True
    else:
        return False
    
Amstronhg(153)
```

W5A18 VIết hàm kiểm tra một số có phải số đối xứng không, xử lý trên dữ liệu kiểu số nguyên.

```python
def sodoixung(n):
    n2 = n
    number = 0
    while n2 % 10 != 0:
        number = number * 10 + n2 % 10
        n2 = n2 // 10
    if number == n:
        return True
    else:
        return False

sodoixung(131)
```

W5A19 VIết hàm kiểm tra một số có phải số kỳ diệu hay không, xử lý trên dữ liệu kiểu số nguyên.

```python
def sokidieu(n):
    s = 0
    for i in range(1, n):
        if n % i == 0:
            s += i
    if s == n:
        return True
    else:
        return False

sokidieu(6)
```

W5A20 Viết hàm detect_number_type(s) trả về 'int' nếu s là số nguyên, 'float' nếu là số thực, ngược lại 'neither'.

```python
def detect_number_type(s):
    if type(s) == int or type(s) == float:
        if s % 1 == 0:
            return "int"
        elif s % 1 != 0:
            return "float"
    else:
        return "neither"

detect_number_type("nbv")
```

W5A24 Viết hàm are_collinear(x1,y1,x2,y2,x3,y3) kiểm tra ba điểm có thẳng hàng hay không.

```python
def are_collinear(x1,y1,x2,y2,x3,y3):
    if (x2 - x1) / (x3 - x1) == (y2 - y1) / (y3 - y1):
        return True
    else:
        return False
    
are_collinear(1,2,2,4,3,6)
```

W5A25:
Viết hàm point_in_quadrilateral(quad, p) kiểm tra điểm p có nằm trong hoặc trên biên của tứ giác quad hay không. Trong đó: quad là một tứ giác (4 đỉnh – 4 điểm), p là một điểm cần kiểm tra. VD: quad = [(0,0), (4,0), (4,3), (0,3)], p = [2, 1]

```python
def point_in_quadrilateral(quad, p):
    v1 = [p[0] - quad[0][0], p[1] - quad[0][1]]
    v2 = [p[0] - quad[1][0], p[1] - quad[1][1]]
    v3 = [p[0] - quad[2][0], p[1] - quad[2][1]]
    v4 = [p[0] - quad[3][0], p[1] - quad[3][1]]
    vecto1 = [quad[1][0] - quad[0][0], quad[1][1] - quad[0][1]]
    vecto2 = [quad[3][0] - quad[0][0], quad[3][1] - quad[0][1]]
    
    tugiac = abs(vecto1[0]*vecto2[1] - vecto2[0]*vecto1[1])
    tamgiac1 = 1/2 * abs(v1[0]*v2[1] - v2[0]*v1[1])
    tamgiac2 = 1/2 * abs(v2[0]*v3[1] - v3[0]*v2[1])
    tamgiac3 = 1/2 * abs(v3[0]*v4[1] - v4[0]*v3[1])
    tamgiac4 = 1/2 * abs(v1[0]*v4[1] - v4[0]*v1[1])

    if tugiac >= (tamgiac1 + tamgiac2 + tamgiac3 + tamgiac4):
        return True
    else:
        return False
    
point_in_quadrilateral([(0,0), (4,0), (4,3), (0,3)], [2, 1])
```

W5A28 Viết hàm tính góc ∠ABC (độ) từ tọa độ A(x1,y1), B(x2,y2), C(x3,y3).

```python
def goc(x1, y1, x2, y2, x3, y3):
    import math
    v1 = [x1 - x2, y1 - y2]
    v2 = [x3 - x2, y3 - y2]
    cos = abs((v1[0]*v2[0] + v1[1]*v2[1])/((v1[0]**2 + v1[1]**2)**0.5 * (v2[0]**2 + v2[1]**2)**0.5))
    return math.acos(cos)*180/math.pi

goc(2,4,5,1,2,1)
```

W5A29 Viết hàm trả về giá trị lớn nhất và nhỏ nhất của ba số a,b,c dưới dạng list [max, min].

```python
def maxmin(a, b, c):
    number = [a, b, c]
    ma = float("-inf")
    mi = float("inf")
    for i in number:
        if i > ma:
            ma = i
        if i < mi:
            mi = i
    return [ma, mi]

maxmin(2, 4, -9)
```

W5A30 Viết hàm tính hệ số nhị thức C(n, k).

```python
def giaithua(n):
    g = 1
    for i in range(1, n + 1):
        g *= i
    return g


def nhithuc(n, k):
    return giaithua(n)/(giaithua(k)*giaithua(n - k))

nhithuc(6, 4)
```
