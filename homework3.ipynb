```python
a = int(input("Nhập số nguyên dương a: "))
b = int(input("Nhập số nguyên dương b: "))

uoc_chung = []
for i in range(1, min(a, b) + 1):
    if a % i == 0 and b % i == 0:
        uoc_chung.append(i)

print("Các ước dương chung của", a, "và", b, "là:", uoc_chung)
```

Bài 10: Nhập vào số nguyên dương a, kiểm tra xem a có phải là số hoàn hảo hay không. 
Số hoàn hảo là số có tổng các ước (ước nhỏ hơn nó) bằng chính nó, VD: số 6, 28… 

```python
a = int(input())

if a <= 0:
    print("khong phai so hoan hao")
else:
    uoc = []
    for i in range(1, a):
        if a % i == 0:
            uoc.append(i)
    
    tong = sum(uoc)
    
    if tong == a:
        print( "⇒ là số hoàn hảo")
    else:
        print( "⇒ không phải số hoàn hảo")
```

Bài 13: Nhập vào số nguyên dương A, tìm n nhỏ nhất sao cho 
1 + 1/2 + 1/3 + 1/4 + ... + 1/n > A

Bài 14: Nhập vào một dãy số nguyên, dừng nhập khi người dùng nhập -1. Sau khi nhập 
xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập. KHÔNG ĐƯỢC dùng mảng


```python
n = int(input())

if n != -1:
    max_val = n
    min_val = n
    
    while True:
        n = int(input())
        if n == -1:
            break
        
        if n > max_val:
            max_val = n
        if n < min_val:
            min_val = n
            
    print(max_val)
    print(min_val)

```

Bài 15: Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số. Làm bằng nhiều 
cách nhất có thể, ko dùng mảng.

```python
n = int(input())
count = 0

if n == 0:
    count = 1
else:
    temp = n
    while temp > 0:
        temp //= 10
        count += 1

print(count)
```

Bài 16: Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số chẵn, bao nhiêu 
chữ số lẻ


```python
n = int(input())

chan = 0
le = 0

if n == 0:
    chan = 1
else:
    temp = n
    while temp > 0:
        chu_so = temp % 10
        if chu_so % 2 == 0:
            chan += 1
        else:
            le += 1
        temp //= 10

print(chan)
print(le)
```

Bài 18: Nhập vào một số nguyên dương n, kiểm tra xem n có phải là số dạng 3^k hay 
không 

```python
n = int(input())

if n <= 0:
    print("False")
else:
    temp = n
    while temp % 3 == 0:
        temp //= 3
    
    if temp == 1:
        print("True")
    else:
        print("False")
```

Bài 19: Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số 
kế tiếp sẽ bằng tổng hai số trước đó. Nhập vào số nguyên dương A, hãy tìm số trong 
dãy số fibonacci lớn nhất nhưng không vượt quá A

```python
A = int(input())

f1 = 1
f2 = 1

if A < 1:
    print("Không có số nào")
elif A == 1:
    print(1)
else:
    while f1 + f2 <= A:
        f_next = f1 + f2
        f1 = f2
        f2 = f_next
    
    print(f2)
```

Bài 20: Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số là số nguyên tố.

```python
n = int(input())
count = 0

temp = n
while temp > 0:
    chu_so = temp % 10
    if chu_so == 2 or chu_so == 3 or chu_so == 5 or chu_so == 7:
        count += 1
    temp //= 10

print(count)
```
